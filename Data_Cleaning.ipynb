## 1. Quantity Ordered

```python
## Let us check the unique values of Quantity Ordered

total_data['Quantity Ordered'].value_counts()
```

So we can see that there are some disprencies in the data and that is the reason we have object type in the data type of the column.

Some issues noticed as of now
1. Quantity Ordered is used as value too in some cases
2. 1 order is having 0 Quantity Ordered
3. 1 order is having 9999 Quantity Ordered


```python
## Let us solve the issue with 0 order and 9999 order

total_data[(total_data['Quantity Ordered']=='0')|(total_data['Quantity Ordered']=='9999')]
```

By looking at the data we can clearly see that these two rows does not make any sense. These rows can be removed from the data


```python
## Removing the above two rows

total_data = total_data.drop(total_data[(total_data['Quantity Ordered']=='0')|(total_data['Quantity Ordered']=='9999')].index).reset_index(drop=True)
```

```python
## We will look into second issue - Quantity Ordered as a row value

#Let us check different values in other columns when row value for Quantity Ordered is Quantity Ordered

for col in total_data.columns:
  if col == 'Quantity Ordered':
    pass
  else:
    print(total_data[total_data['Quantity Ordered']=='Quantity Ordered'][col].unique())
```

So it is clear that these values have been in the dataset due to some technical or human error. These were not meant to be here. Removing again these value does not have any impact on data quality.

```python
## Removing the above two rows

total_data = total_data.drop(total_data[total_data['Quantity Ordered']=='Quantity Ordered'].index).reset_index(drop=True)
```

Apart from NULL values we can see there are three major issues

```python
total_data[(total_data['Price Each']=='$149.99')|(total_data['Price Each']=='$11.95')]
```
